```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
#read in csv file
data_analyst_file = "Resources/DataAnalyst.csv"
analyst_df = pd.read_csv(data_analyst_file)
analyst_df.head()
```

```python
# Create a filtered dataframe from specific columns
data_cols = ['Unnamed: 0','Job Title', 'Salary Estimate', 'Job Description', 
            'Rating', 'Company Name', 'Location', 'Size', 
            'Founded', 'Type of ownership', 'Industry', 'Sector', 'Revenue']
analyst_transformed= analyst_df[data_cols].copy()
```

```python
# Rename the column headers
analyst_transformed = analyst_transformed.rename(columns={'Unnamed: 0': 'id',
                                                          'Salary Estimate': 'Glass Door Salary Estimate'})

# Set index
analyst_transformed.set_index("id", inplace=True)

analyst_transformed.head()
```

```python
#filter 2 major cities - Los Angeles, CA & New York, NY
la_ny_df = analyst_transformed.loc[(analyst_transformed['Location'] == 'Los Angeles, CA')|\
                                   (analyst_transformed['Location'] == 'New York, NY')].copy()

la_ny_df.head()
```

```python
#Clean up some of the columns by eliminating unwanted text
la_ny_df['Glass Door Salary Estimate']=la_ny_df['Glass Door Salary Estimate']\
    .str.replace("(Glassdoor est.)", "", regex=False)

la_ny_df['Revenue']=la_ny_df['Revenue']\
    .str.replace("Unknown / Non-Applicable", "Unknown", regex=False)
```

```python
#reaplacing unknown values w/ "unknown"
la_ny_df.replace(to_replace = -1, value = "Unknown", inplace = True)
la_ny_df.replace(to_replace = "-1", value = "Unknown", inplace = True)
la_ny_df
```

```python
#show hoy many job titles are in Los Angeles, CA & New York, NY   
la_ny_group = la_ny_df.groupby('Location').agg({'Job Title': ['count']})
print(la_ny_group)
```

```python
#create the data base connection
connection_string = "postgres:postgres@localhost:5432/Data_Jobs"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#confrim table
engine.table_names()
```

```python
#load dataframe into database
la_ny_df.to_sql(name='la_ny_table', con=engine, if_exists='append', index=True)
```
